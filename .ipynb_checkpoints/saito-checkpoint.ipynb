```python
import numpy as np
import glob
import matplotlib.pyplot as plt
import pandas as pd
from scipy import interpolate
from scipy import integrate
```

```python
def read_nihhonkoudentxt(filename=''):
    '''日本光電のデータを読み込み，マーカーを検出する関数'''
    df = pd.read_table(filename, encoding='shift-jis', sep='\t', header=13) #生データ読み込み
    df = np.array(df)
    mark = []
    for i in range(df.shape[0]):
        if df[i, 1] is not np.nan:
          mark.append(i)  # marker index list
          if len(mark)==20: break

    return np.array(df), np.array(mark)
```

```python
def crop_data(data, mark):
     # 心電の取り出し
    rest0 = np.array(data[mark[m]:mark[m]+int(resttime*fs)])[:,0]
    task = np.array(data[mark[m]+int(resttime*fs):mark[m]+int(resttime*fs)+int(tasktime*fs)])[:,0]
    rest1 = np.array(data[mark[m]+int(resttime*fs)+int(tasktime*fs):mark[m]+int(resttime*fs)+int(tasktime*fs)+(resttime*fs)])[:,0]
    
    return rest0, task, rest1
```

```python
def calculate_lfhf(trial):
    # RRIの検出
    RRI = []
    row = 1
    rowcount = -1000
    count = 0
    RR = []

    trial_times = np.arange(0, len(trial), 1)
    #R波の検出
    peak_index_list = []
    while row <= (len(trial)-2):
        if trial[row] >= trial[row-1] and trial[row] >= trial[row+1]:
            if trial[row] >= HRmin and (row-rowcount) >= tmin:
                if count == 0:
                    RR.append(row)
                    peak_index_list.append(row)
                    count += 1
                    rowcount = row
                else:
                    RR.append(row)
                    peak_index_list.append(row)
                    RRI.append(RR[count]-RR[count-1])
                    count += 1
                    rowcount = row
        row += 1

    RRI = np.array(RRI)/fs # msecに変換    
    RR = np.array(RR[1:])/fs
    
    

    # linear interpolation
    hokan = interpolate.interp1d(RR, RRI, kind='cubic')
    RRnew = np.arange(RR[0], RR[-1], 0.001)
    RRInew = hokan(RRnew)
    print('resampled_rr_intervals -> {}'.format(RRInew.shape))
    print('resampled_rr_intervals -> {}'.format(RRInew))

    # calculate power spectrum density
    F = np.fft.fft(RRInew)
    Amp = np.abs(F)
    Pow = Amp ** 2
    
    print('amplitudes shape -> {}'.format(Amp.shape))
    print('amplitudes -> {}'.format(Amp))
    print('power shape -> {}'.format(Pow.shape))
    print('power -> {}'.format(Pow))
    freq = np.fft.fftfreq(RRInew.size, d=0.001)
    freq = freq[:int(freq.shape[0]/2)]

    print('frequencies shape -> {}'.format(freq.shape))
    print('frequencies -> {}'.format(freq[:20]))
    # integrate
    def PowIntegrate(x0, x1):
        flag0 = np.where(freq>=x0)[0][0]
        print('np.where(freq>=x0) -> {}'.format(np.where(freq>=x0)))
        flag1 = np.where(freq<=x1)[0][-1]
        print('flag0 -> {}'.format(flag0))
        print('flag1 -> {}'.format(flag1))
        print('powers -> {}'.format(Pow[flag0:flag1+1]))
        # PowFunc = lambda j: Pow[j]
        return integrate.simps(Pow[flag0:flag1+1], freq[flag0:flag1+1])

    lf = PowIntegrate(0.05, 0.15)
    hf = PowIntegrate(0.15, 0.4)
    print('low -> {}'.format(np.sum(lf)))
    print('high -> {}'.format(np.sum(hf)))
    print('lhhf -> {}'.format(lf/hf))
    plt.plot(trial_times, trial, '--', trial_times[peak_index_list], trial[peak_index_list], 'x')
    plt.plot(RR, RRI, '.', RRnew, RRInew, 'x')
    plt.show()
    plt.plot(Pow[1:])
    plt.show()
    
    return lf/hf
```

```python
# パラメータの設定
data_dir = './nihonkoden-ecg'
ecg_filepath_list = glob.glob('{}/*txt'.format(data_dir))
filename = ecg_filepath_list[0]
resttime = 45 #sec
tasktime = 30 #sec
fs = 1000 #サンプリング周期(1000HZ)
HRmin = 0.4 # HRの下限値1.0s(閾値)
tmin = int(0.5*fs) # RRIの最小時間0.5sec(閾値)

# データの読み込み
data, mark = read_nihhonkoudentxt(filename)

LFHF = []
event = []

for m in range(len(mark)):
    lfhf = []
    for t in crop_data(data, mark)[1:2]:
        lfhf.append(calculate_lfhf(t))
    
    LFHF.append(lfhf)
    event.append(m)
    print(m, lfhf)
```

```python
LFHF = np.array(LFHF)
LFHF = pd.DataFrame({'trial':event, 'task':LFHF[:,0]})
```
